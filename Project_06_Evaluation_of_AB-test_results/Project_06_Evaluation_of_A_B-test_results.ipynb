## Приоритизация гипотез

Загрузим необходимые библиотеки

```python
import pandas as pd
import numpy as np
import datetime as dt
import scipy.stats as stats
from matplotlib import pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

Загрузим и изучим данные датасета с гипотезами. Сохраним в переменную `hypothesis`

```python
hypothesis = pd.read_csv('/datasets/hypothesis.csv') 
pd.set_option('display.max_colwidth', None)
hypothesis
```

В файле мы видим 9 гипотез по увеличению выручки интернет-магазина с указанными параметрами Reach, Impact, Confidence, Effort. Гипотезы оценивались по 10-балльной шкале.

Описание данных: 
- Hypothesis — краткое описание гипотезы;
- Reach — охват пользователей;
- Impact — влияние на пользователей;
- Confidence — уверенность в гипотезе;
- Efforts — затраты ресурсов на проверку гипотезы по 10-балльной шкале. Чем больше значение Efforts, тем дороже проверка гипотезы.

Задача отсортировать гипотезы по приоритету.  Для этого используем методы **ICE** и **RICE**. Сравним результаты и сделаем выводы.

**Применим фреймворк ICE для приоритизации гипотез.** Отсортируем их по убыванию приоритета.

```python
hypothesis['ICE'] = round(hypothesis['Impact'] * hypothesis['Confidence'] / hypothesis['Efforts'])

hypothesis[['Hypothesis', 'ICE']].sort_values(by='ICE', ascending=False)
```

Наиболее перспективные гипотезы по ICE: 8, 0, 7

**Применим фреймворк RICE для приоритизации гипотез.** Отсортируем их по убыванию приоритета.

```python
hypothesis['RICE'] = hypothesis['Reach'] * hypothesis['Impact'] * hypothesis['Confidence'] / hypothesis['Efforts']

hypothesis[['Hypothesis', 'RICE']].sort_values(by='RICE', ascending=False)
```

Наиболее перспективные гипотезы по RICE: 7, 2, 0 и 6

Выведем сравнительную таблицу с двумя параметрами вместе для наглядности.

```python
hypothesis[['Hypothesis', 'ICE', 'RICE']].sort_values(by='ICE', ascending=False)
```

**Вывод**

Мы выявили наиболее перспективные гипотезы, которые следует проверит в первую очередь, это гипотезы с номерами 0, 2, 7, 8.
Однако в большем приоритете по методу ICE была гипотеза номер 8, а по методу RICE - номер 7. Так получилось из-за параметра охвата, т.к. добавление формы подписки на страницы затронет большее количество пользователей, чем акция для именинников. 
То же самое можно сказать о гипотезе 2, у нее низкая оценка влияния, но высокий охват, поэтому по методу RICE она оказалась на втором месте.

### Сбор кумулятивных данных

Для дальнейшего анализа и построения графиков нам нужно собрать кумулятивные данные. Объявим датафрейм `data` и передадим необходимые столбцы. 

```python
visitorsADaily = visitors[visitors['group'] == 'A'][['date', 'visitors']]
visitorsADaily.columns = ['date', 'visitorsPerDateA']

visitorsACummulative = visitorsADaily.apply(
    lambda x: visitorsADaily[visitorsADaily['date'] <= x['date']].agg(
        {'date': 'max', 'visitorsPerDateA': 'sum'}
    ),
    axis=1,
)
visitorsACummulative.columns = ['date', 'visitorsCummulativeA']

visitorsBDaily = visitors[visitors['group'] == 'B'][['date', 'visitors']]
visitorsBDaily.columns = ['date', 'visitorsPerDateB']

visitorsBCummulative = visitorsBDaily.apply(
    lambda x: visitorsBDaily[visitorsBDaily['date'] <= x['date']].agg(
        {'date': 'max', 'visitorsPerDateB': 'sum'}
    ),
    axis=1,
)
visitorsBCummulative.columns = ['date', 'visitorsCummulativeB']

ordersADaily = (
    orders[orders['group'] == 'A'][['date', 'transactionId', 'visitorId', 'revenue']]
    .groupby('date', as_index=False)
    .agg({'transactionId': pd.Series.nunique, 'revenue': 'sum'})
)
ordersADaily.columns = ['date', 'ordersPerDateA', 'revenuePerDateA']

ordersACummulative = ordersADaily.apply(
    lambda x: ordersADaily[ordersADaily['date'] <= x['date']].agg(
        {'date': 'max', 'ordersPerDateA': 'sum', 'revenuePerDateA': 'sum'}
    ),
    axis=1,
).sort_values(by=['date'])
ordersACummulative.columns = [
    'date',
    'ordersCummulativeA',
    'revenueCummulativeA',
]

ordersBDaily = (
    orders[orders['group'] == 'B'][['date', 'transactionId', 'visitorId', 'revenue']]
    .groupby('date', as_index=False)
    .agg({'transactionId': pd.Series.nunique, 'revenue': 'sum'})
)
ordersBDaily.columns = ['date', 'ordersPerDateB', 'revenuePerDateB']

ordersBCummulative = ordersBDaily.apply(
    lambda x: ordersBDaily[ordersBDaily['date'] <= x['date']].agg(
        {'date': 'max', 'ordersPerDateB': 'sum', 'revenuePerDateB': 'sum'}
    ),
    axis=1,
).sort_values(by=['date'])
ordersBCummulative.columns = [
    'date',
    'ordersCummulativeB',
    'revenueCummulativeB',
]

data = (
    ordersADaily.merge(ordersBDaily, how='left')
    .merge(ordersACummulative, how='left')
    .merge(ordersBCummulative, how='left')
    .merge(visitorsADaily, how='left')
    .merge(visitorsBDaily, how='left')
    .merge(visitorsACummulative, how='left')
    .merge(visitorsBCummulative, how='left')
)

data.head()
```

Описание столбцов таблицы `data`:
- date — дата;
- ordersPerDateA — количество заказов в выбранную дату в группе A;
- revenuePerDateA — суммарная выручка в выбранную дату в группе A;
- ordersPerDateB — количество заказов в выбранную дату в группе B;
- revenuePerDateB — суммарная выручка в выбранную дату в группе B;
- ordersCummulativeA — суммарное число заказов до выбранной даты включительно в группе A;
- revenueCummulativeA — суммарная выручка до выбранной даты включительно в группе A;
- ordersCummulativeB — суммарное количество заказов до выбранной даты включительно в группе B;
- revenueCummulativeB — суммарная выручка до выбранной даты включительно в группе B;
- visitorsPerDateA — количество пользователей в выбранную дату в группе A;
- visitorsPerDateB — количество пользователей в выбранную дату в группе B;
- visitorsCummulativeA — количество пользователей до выбранной даты включительно в группе A;
- visitorsCummulativeB — количество пользователей до выбранной даты включительно в группе B.

**Вывод**

Мы обработали исходные данные, выполнили необходимые подсчеты и собрали таблицу `data`, которую используем для дальнйшего анализа и построения графиков.

### График кумулятивной выручки по группам.

Построим график кумулятивной выручки по группам. Используем данные таблицы `data`

```python
# Строим график выручки группы А
plt.plot(data['date'], data['revenueCummulativeA'], label='A')

# Строим график выручки группы B
plt.plot(data['date'], data['revenueCummulativeB'], label='B')

plt.legend()
plt.ticklabel_format(axis='y', style='plain')
plt.title('Кумулятивная выручка по группам')
plt.ylabel('Выручка, руб.')
plt.xticks(rotation=45, ha='right')
plt.grid(True);
```

Выручка постепенно увеличивается в течение всего теста, это хорошо. Группа В показывает стабильно лучшие результаты, чем группа А. На графике группы В после 17-го числа виден резкий скачок. Это могут быть очень дорогие заказы, либо очень большое количество заказов. Нужно проверить данные на наличие выбросов.

### График кумулятивного среднего чека по группам

```python
plt.plot(data['date'], data['revenueCummulativeA']/data['ordersCummulativeA'], label='A')
plt.plot(data['date'], data['revenueCummulativeB']/data['ordersCummulativeB'], label='B')

plt.legend()
plt.title('Кумулятивный средний чек по группам')
plt.ylabel('Средний чек, руб.')
plt.xticks(rotation=45, ha='right')
plt.grid(True);
```

На график группы В как и ожидалось виден резкий всплеск. Делать выводы по такому графику рано, т.к. кумулятивное значение среднего чека колеблется и если не учитывать выбросы в группе В, картина может быть не однозначной.

### График относительного изменения кумулятивного среднего чека группы B к группе A

```python
# cтроим отношение средних чеков
plt.plot(data['date'], (data['revenueCummulativeB']/data['ordersCummulativeB']) /
         (data['revenueCummulativeA']/data['ordersCummulativeA']) 
         -1)

plt.axhline(y=0, color='black', linestyle='--')
plt.title('Относительное изменение кумулятивного среднего чека группы B к группе A')
plt.xticks(rotation=45, ha='right')
plt.grid(True);
```

На графике также наблюдаются резкие скачки, говорящие о наличии выбросов. На первый взгляд показатели группы В лучше, но делать окончательные выводы по таким данным преждевременно.

### График кумулятивной конверсии по группам

```python
plt.plot(data['date'], data['ordersCummulativeA']/data['visitorsCummulativeA'], label='A')
plt.plot(data['date'], data['ordersCummulativeB']/data['visitorsCummulativeB'], label='B')

plt.legend()
plt.title('Кумулятивная конверсия по группам')
plt.ylabel('Уровень конверсии')
plt.xticks(rotation=45, ha='right')
plt.axis([dt.datetime(2019, 8, 1), dt.datetime(2019, 8, 31), 0.02, 0.05])
plt.grid(True);
```

На график добавили масштаб, чтобы сгладить колебания. Видно, что конверсия группы почти сразу В обогнала группу А и зафиксировалась примерно на уровне 0,035. Конверсия группы А просела и также зафиксировалась.

### График относительного изменения кумулятивной конверсии группы B к группе A

```python
plt.plot(data['date'], (data['ordersCummulativeB']/data['visitorsCummulativeB']) /
         (data['ordersCummulativeA']/data['visitorsCummulativeA']) -1)

plt.title('Относительный прирост конверсии группы B относительно группы A')
plt.axhline(y=0, color='black', linestyle='--')
plt.xticks(rotation=45)
plt.axhline(y=0.140, color='grey', linestyle='--')
plt.axis([dt.datetime(2019, 8, 1), dt.datetime(2019, 8, 31), -0.2, 0.4]) 
plt.grid(True);
```

### Статистическая значимость различий в конверсии между группами по «сырым» данным

Создадим переменные `ordersByUsersA` и `ordersByUsersB`, где для пользователей, совершивших хотя бы 1 заказ, будет указано число заказов.

```python
ordersByUsersA = (
    orders[orders['group'] == 'A']
    .groupby('visitorId', as_index=False)
    .agg({'transactionId': pd.Series.nunique})
)
ordersByUsersA.columns = ['userId', 'orders']

ordersByUsersB = (
    orders[orders['group'] == 'B']
    .groupby('visitorId', as_index=False)
    .agg({'transactionId': pd.Series.nunique})
)
ordersByUsersB.columns = ['userId', 'orders'] 

ordersByUsersA.head(), ordersByUsersB.head()
```

Создадим переменные `groupA` и `groupB`, в которых пользователям с заказами будет соответствовать число заказов пользователя. А пользователям без заказов — нули.

```python
groupA = pd.concat([ordersByUsersA['orders'],pd.Series(0, index=np.arange(data['visitorsPerDateA'].sum() - len(ordersByUsersA['orders'])), name='orders')],axis=0)
groupB = pd.concat([ordersByUsersB['orders'],pd.Series(0, index=np.arange(data['visitorsPerDateB'].sum() - len(ordersByUsersB['orders'])), name='orders')],axis=0)
```

Посчитаем статистическую значимость различия конверсии по результатам А/В теста, применим тест Манна-Уитни. Параметр pvalue = 0.05

```python
"{0:.5f}".format(stats.mannwhitneyu(groupA, groupB)[1])
```

Посмотрим на относительное различие в конверсии между группами.

```python
"{0:.5f}".format(groupB.mean() / groupA.mean() - 1)
```

p-value = 0.017, это меньше, чем пороговое значение 0.05. Значит по "сырым" данным можно сделать вывод, что в конверсии групп А и В есть статистически значимые различия. 

Относительная разница между группами - 14%, т.е. группа В показывает лучший результат.

### Статистическая значимость различий в среднем чеке заказа между группами по «сырым» данным

```python
'{0:.5f}'.format(stats.mannwhitneyu(orders[orders['group']=='A']['revenue'], orders[orders['group']=='B']['revenue'])[1])
```

Определим и относительную разницу между показателями среднего чека

Относительное различие между группами целых 26%. 

### Статистическая значимость различий в конверсии между группами по «очищенным» данным

Избавимся от аномальных значений в группах. Удалим пользователей, сделавших больше 3-х заказов и тех, кто сделал заказы на сумму больше 30 000 рублей. 

Для начала создадим таблицу `abnormalUsers` с аномальными пользователями, сделав срезы по нашим условиям.

```python
usersWithManyOrders = pd.concat(
    [
        ordersByUsersA[ordersByUsersA['orders'] > orders_percentile[1]]['userId'],
        ordersByUsersB[ordersByUsersB['orders'] > orders_percentile[1]]['userId'],
    ],
    axis=0,
)
usersWithExpensiveOrders = orders[orders['revenue'] > cost_percentile[1]]['visitorId']
abnormalUsers = (
    pd.concat([usersWithManyOrders, usersWithExpensiveOrders], axis=0)
    .drop_duplicates()
    .sort_values()
)
display(abnormalUsers.head(5))
len(abnormalUsers)
```

Найдено 15 пользователей с аномальной активностью. 

Создадим переменные `groupAFiltered` и `groupBFiltered`, в которых будут очищенные данные о заказах, без аномальных пользователей.

```python
groupAFiltered = pd.concat(
    [
        ordersByUsersA[
            np.logical_not(ordersByUsersA['userId'].isin(abnormalUsers))
        ]['orders'],
        pd.Series(
            0,
            index=np.arange(
                data['visitorsPerDateA'].sum() - len(ordersByUsersA['orders'])
            ),
            name='orders',
        ),
    ],
    axis=0,
)

groupBFiltered = pd.concat(
    [
        ordersByUsersB[
            np.logical_not(ordersByUsersB['userId'].isin(abnormalUsers))
        ]['orders'],
        pd.Series(
            0,
            index=np.arange(
                data['visitorsPerDateB'].sum() - len(ordersByUsersB['orders'])
            ),
            name='orders',
        ),
    ],
    axis=0,
) 
```

Определим p-value для сравнения конверсии между очищенными группами.

```python
'{0:.5f}'.format(stats.mannwhitneyu(groupAFiltered, groupBFiltered)[1])
```

Посмотрим на относительное различие в конверсии между группами по очищенным данным

```python
'{0:.3f}'.format(groupBFiltered.mean()/groupAFiltered.mean()-1)
```

Значение p-value практически не изменилось, оно также меньше 0.05, как и в случае с "сырами" данными. Можно сказать, что между конверсиями групп А и В есть статистически значимые различия. 

Относительное различие в конверсии между группами по очищенным данным увеличилось примерно на 1%, группа В показывает на 15% лучший результат, чем группа А.

### Статистическая значимость различий в среднем чеке заказа между группами по «очищенным» данным

Найдем p-value для сравнения средних чеков между очищенными группами. Используем критерий Манна-Уитни.

```python
'{0:.5f}'.format(
        stats.mannwhitneyu(
            orders[
                np.logical_and(
                    orders['group'] == 'A',
                    np.logical_not(orders['visitorId'].isin(abnormalUsers)),
                )
            ]['revenue'],
            orders[
                np.logical_and(
                    orders['group'] == 'B',
                    np.logical_not(orders['visitorId'].isin(abnormalUsers)),
                )
            ]['revenue'],
        )[1]
    )
```

Определим относительное различие в средних чеках по очищенным данным

```python
print(
    "{0:.3f}".format(
        orders[
            np.logical_and(
                orders['group'] == 'B',
                np.logical_not(orders['visitorId'].isin(abnormalUsers)),
            )
        ]['revenue'].mean()
        / orders[
            np.logical_and(
                orders['group'] == 'A',
                np.logical_not(orders['visitorId'].isin(abnormalUsers)),
            )
        ]['revenue'].mean()
        - 1
    )
) 
```

Значение p-value по очищенным данным стало еще дальше от порогового. Можно сказать, что статистически значимых различий в средних чеках между тестируемыми группами нет.

Выбросы оказали очень сильное влияние на показатели относительного различия по среднему чеку. "Сырые" данные показывали прирост в 26%, по очищенным видим, что группа В наоборот, на 0.6% проигрывает группе А. Такое различие в данном случае не является статистически значимым.
